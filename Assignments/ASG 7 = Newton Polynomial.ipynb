```python
import pandas as pd
f = pd.read_excel('Sir x4 Data.xlsx')
display(f)
```

```python
n = len(f)
x = f.iloc[:,0].tolist()
y = f.iloc[:,1].tolist()
```

```python
x0 = float(input('Enter a probe point: '))
```

```python
def NDDM(n,x,y,x0):
  p = [0]*n
  a = [0]*n
  p[0] = y[0]
  h = 1
  for k in range(1,n):
    pro = 1
    for j in range(k):
      pro*=(x[k]-x[j])
    a[k] = (y[k] - p[k-1])/pro
    h*=(x0-x[k-1])
    p[k] = p[k-1] + (a[k]*h)
    #print(p[k])
  a = p[-1]
  b = p[int(n/2)]
  return (a,b)
```

```python
NDDM(n,x,y,x0)
```

Value of 6 degree polynomial function p(x) passing through 7 (x,y) points at x = 0.55 is ~ 0.09<br>
Value of 3 degree polynomial function p(x) passing through 7 (x,y) points at x = 0.55 is ~ 0.1

```python
import numpy as np
x0_range = np.arange(0,1.25,0.05).tolist()
y_int = [NDDM(n,x,y,xi)[0] for xi in x0_range]
y_half_int = [NDDM(n,x,y,xi)[1] for xi in x0_range]
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.plot(x0_range,y_int)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['Data Point Curve','Interpolated Curve'])
```

```python
y0_range = [xi**4 for xi in x0_range]
e_int = [y-f for (y,f) in zip(y_int,y0_range)]
e_half_int = [y-f for (y,f) in zip(y_half_int,y0_range)]
plt.plot(x0_range,e_int)
plt.plot(x0_range,e_half_int)
plt.xlabel('x')
plt.ylabel('Error')
plt.legend(['6 degree','3 degree'])
```
