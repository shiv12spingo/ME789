```python
import math
```

```python
# Reading inputs for the problem
a_d = float(input("Smallest Angle in degrees = "))
a = math.radians(a_d)
b_d = float(input("Largest Angle in degrees = "))
b = math.radians(b_d)
R = float(input("Range = "))
V = float(input("Initial Speed = "))
tol = float(input("Tolerance = "))
N = int(input("Max no of iterations = "))
```

```python
# Defining function for root finding equation
# y as a function of x given velocity v and angle theta
def Trajectory_Function(x,v,theta):
  t = math.tan(theta)
  c = math.cos(theta)
  y = (t*x)-(0.5*9.81*((x/(v*c))**2))
  return y
```

```python
# Defining function for Bisection method
def Bisection(a,b,R,V,tol,N):
  iter = 1
  FA = Trajectory_Function(R,V,a)
  while (iter<N):
    p = a + ((b-a)/2) # finding midpoint of a and b
    FP = Trajectory_Function(R,V,p)
    if (FP==0 or ((b-a)/2)<tol):
      print("Root = "+str(round(math.degrees(p),3))+" degrees found in "+str(iter)+" iterations")
      break
    iter+=1
    if (FA*FP)>0:
      a=p
      FA=FP
    else: 
      b=p
    if iter==N:
      print("Maximum Iterations Reached")
```

```python
# Run Bisection Method and display output
Bisection(a,b,R,V,tol,N)
```

```python
# Displaying the projectile trajectory
Theta = math.radians(5.658) # from above output
import numpy as np
import matplotlib.pyplot as plt
x = (np.linspace(0,R,100)).tolist()
y = [Trajectory_Function(item,V,Theta) for item in x]
plt.plot(x,y)
plt.title("Trajectory Profile")
plt.xlabel("x")
plt.ylabel("y")
```
