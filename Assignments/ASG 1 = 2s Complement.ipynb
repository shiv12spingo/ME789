```python
# Taking all inputs and assigning variables
n = int(input("Number of Bits = "))
number = int(input("Enter Integer = "))
```

```python
# Converting absolute value of decimal 'number' to 'n' bit binary string
binary = format(abs(number),f'0{n}b')
```

```python
if number>=0:
  # Twos complement of a +ve decimal 'number' is its 'n' bit binary form
  print('Twos Complement = '+binary)
else:
  # Flipping all bits
  flip1 = binary.replace('1','2')
  flip2 = flip1.replace('0','1')
  flip = flip2.replace('2','0')
  # Converting binary 'flip' to decimal and adding 1
  decimal = int(flip,2)
  new_number = decimal+1
  # Converting decimal 'new_number' to 'n' bit binary string
  new_binary = format(new_number,f'0{n}b')
  print('Twos Complement = '+new_binary)
```

```python
twos_complement = input('Twos Complement = ')
no_of_bits = len(twos_complement)
decimal = int(twos_complement,2)
if twos_complement[0]=='0':
  print('Your Integer = '+str(decimal))
else:
  new_decimal = (2**no_of_bits)-decimal
  print('Your Integer = '+str(-new_decimal))
```
