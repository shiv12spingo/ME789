```python
import numpy as np
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import mean_squared_error
from statsmodels.tools.eval_measures import rmse
import warnings
from pandas.util._decorators import Appender
#from pmdarima import auto_arima

warnings.filterwarnings("ignore")
```

```python
from pmdarima import auto_arima
```

```python
df.Month_Year = pd.to_datetime(df.Month_Year)
df = df.set_index("Month_Year")
df.head()
```

```python
auto_arima(df['Rainfall'], seasonal=True, m=12,max_p=7, max_d=5,max_q=7, max_P=4, max_D=4,max_Q=4).summary()
```

```python
train_data = df[:1000]
test_data = df[1000:]
arima_model = SARIMAX(train_data['Rainfall'], order = (2,1,1), seasonal_order = (4,0,3,12))
arima_result = arima_model.fit()
arima_result.summary()
```

```python
arima_pred = arima_result.predict(start = len(train_data)+12, end = len(df)-1, typ="levels").rename("ARIMA Predictions")
arima_pred
```

```python
A = arima_pred
```

```python
test_data['Rainfall'][12:].plot(figsize = (16,5), legend=True)
arima_pred.plot(legend = True);
```

```python
arima_rmse_error = rmse(test_data['Rainfall'][12:], arima_pred)
arima_mse_error = arima_rmse_error**2
mean_value = df['Rainfall'].mean()

print(f'MSE Error: {arima_mse_error}\nRMSE Error: {arima_rmse_error}\nMean: {mean_value}')
```

```python
test_data['ARIMA_Predictions'] = arima_pred
```

```python
def mean_absolute_percentage_error(y_true, y_pred):

  y_true, y_pred = np.array(y_true), np.array(y_pred)

  return np.mean(np.abs((y_true - y_pred) / y_true))*100

mean_absolute_percentage_error(test_data['Rainfall'][12:],arima_pred)
```
